### Konlpy란

한글 자연어 처리를 위한 패키지로 아래의 기능을 가진다.

형태소 분석: 한국어 텍스트를 형태소 단위로 분리하고 품사를 태깅한다.

토큰화: 문장을 단어나 형태소 단위로 나누는 기능을 제공한다.



주요 기능:

morphs(): 텍스트에서 형태소를 추출한다.

nouns(): 텍스트에서 명사를 추출한다.

pos(): 텍스트의 품사 정보를 태깅한다.

```python
from konlpy.tag import Okt #konlpy의 Okt 형태소 분석기를 받아온 후 문장이 들어오면 어근을 분리하는 함수 생성, konlpy패키지 별도 설치 필요


okt=Okt() 
def okt_tokenizer(sentence):
    okt_token=okt.morphs(sentence) #.morphs로 텍스트를 형태소 단위로 나눈다
    return okt_token

okt_tokenizer(interview_data.iloc[0,0])

```

### 불용어 처리란
언어 처리에서 의미를 가지지 않거나 분석에 큰 도움이 되지 않는 일반적인 단어들을 말한다. ex) 그 , 저, 입니다.

불용어의 주요 특징:

텍스트에 빈번하게 등장하지만 문맥적으로 중요한 의미를 갖지 않습니다.

텍스트 분석 시 제거되어 처리 효율성을 높이고 분석의 정확도를 개선합니다.

작업의 목적과 언어에 따라 불용어 목록이 달라질 수 있습니다.

```python
stop_words=pd.read_csv('korean_stopwords.txt') #불용어 처리를 위해 korean_stopwords.txt의 파일을 불러온다.
```

```python
stop_words_interview=stop_words['불용어'].tolist() #불용어 리스트를 생성
stop_words_interview
```

Tf-idf는 Term frequency-inverse document frequency의 약자.

TF(Term Frequency)는 특정 문서 내에서 단어가 얼마나 자주 등장하는지를 나타내는 지표

- 한 문서 내에서 많이 등장하는 단어는 중요!
(단순 빈도로 계산)

IDF(Inverse Document Frequency)는 단어의 전체 문서 집합에서 얼마나 희귀한 단어인지를 나타내는 지표

- 한 문서 내에서 희소하게 등장하는 단어는 다른 문서와 구별해주므로 중요!
(전체 문서에서 해당 단어가 포함된 문서의 개수를 나눈뒤 로그)
 
TF-IDF는 TF와 IDF를 곱하여 단어의 중요도를 계산. 특정 문서에서 자주 등장하면서 전체 문서에서는 희귀한 단어일수록 TF-IDF 값이 높아져 이를 통해 텍스트 데이터에서 특정 단어의 중요도를 평가하고, 문서 간의 단어 비교와 검색에서 사용할 수 있다.

### 주요 파라미터

ngram_range: 단어 시퀀스의 범위를 지정한다.

max_features: 벡터화 과정에서 고려할 최대 특성(단어) 수를 지정한다. 

use_idf: IDF 가중치를 적용할지 여부를 결정한다. 기본값은 True이다.

sublinear_tf: TF 값에 대해 1+log(tf) 스케일링을 적용할지 결정한다.

min_df, max_df: 문서 빈도의 최소값과 최대값을 지정하여 너무 희귀하거나 흔한 단어를 제외할 수 있다.

```python
from sklearn.feature_extraction.text import TfidfVectorizer #Tfidf벡터화를 통해 각각의 어근을 피쳐로 갖는 희소행렬을 생성, stop_words 파라미터를 통해 불용어 제거


tfidf=TfidfVectorizer(tokenizer=okt_tokenizer,ngram_range=(1,2),stop_words=stop_words_interview ,lowercase=False,) #파라미터는 추후 모델 성능을 비교하며 조정
tfidf.fit(question_cleasinged)
tfidf_interview_data=tfidf.transform(question_cleasinged)
tfidf_interview_data
```

### 위 과정을 통해 tfidf벡터화된 행렬을 반환하는 함수 생성

```python
def get_tfidf(data_file,sentence_column,stop_words_file): #data_file=인터뷰 데이터 파일,sentence_column=텍스트 데이터에 해당하는 컬럼 인덱스,stop_words_file=불용어 파일
    
    interview_data=pd.read_excel(data_file)
    question_cleasinged=interview_data.iloc[:,sentence_column].values.astype('str')
    
    stop_words=pd.read_csv(stop_words_file)
    stop_words_interview=stop_words['불용어'].tolist()
    
    tfidf=TfidfVectorizer(tokenizer=okt_tokenizer,ngram_range=(1,2),max_df=1.0,min_df=0.0, stop_words=stop_words_interview,lowercase=False)
    tfidf.fit(question_cleasinged)
    tfidf_interview_data=tfidf.transform(question_cleasinged)
    tfidf_interview_data
    
    return tfidf_interview_data
```

```python
from sklearn.preprocessing import LabelEncoder

category=interview_data['Category'].values
# LabelEncoder를 객체로 생성한 후 , fit( ) 과 transform( ) 으로 label 인코딩 수행. 
encoder = LabelEncoder()
encoder.fit(category)#학습
category_encoded = encoder.transform(category)#숫자로 변환
print('인코딩 변환값:',category_encoded)
```

```python
print('인코딩 클래스:',encoder.classes_)
```

```python
print('디코딩 원본 값:',encoder.inverse_transform([0,1,2,3,4]))
```

### 위 과정을 통해 인코딩 반환값을 배열로 반환하는 함수 지정

```python
def category_encoding(df, category_index): #입력값으로 데이터 프레임과 인코딩할 컬럼의 인덱스를 받는다.
    category=df.iloc[:,category_index].values
    encoder = LabelEncoder()
    encoder.fit(category)
    category_encoded = encoder.transform(category)
    return category_encoded
```

```python
interview_data['category_encoded']=category_encoded
```
